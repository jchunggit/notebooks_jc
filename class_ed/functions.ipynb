```python
def totalby(df,col1,col2):
        import pandas as pd
        df_f = pd.crosstab(df[col1], df[col2])
        df_f['total'] = df_f.sum(axis = 1)
        df_f = df_f.sort_values('total', ascending = False)
        df_f = df_f[[True, False, 'total']]
        df_f.drop("total", axis = 1, inplace = True) 
        return(df_f)

def var2auto(df_f, title, xlab, ylab):
        import pandas as pd
        import matplotlib.pyplot as plt
        x_vals = df_f.index.nunique()
        t_c = df_f.sum(axis = 1)
        fig, ax = plt.subplots()
        df_f.plot(kind='bar', stacked=True, figsize=(8, 6), 
                  rot=0, xlabel=xlab, ylabel=ylab, ax = ax)
        ax.legend(fontsize=12)
        for i in range(0,x_vals):
                ax.text(i, df_f[1][i]/2, 
                        f'{df_f[1][i]:.0f}\n{df_f[1][i] / t_c[i] * 100.0:.1f} %', 
                        horizontalalignment='center',
                        verticalalignment='center')
                ax.text(i, t_c[i]-df_f[0][i]/2, 
                        f'{df_f[0][i]:.0f}\n{df_f[0][i] / t_c[i] * 100.0:.1f} %', 
                        horizontalalignment='center',
                        verticalalignment='center')
                ax.text(i, t_c[i], 
                        f'{t_c[i]:.0f}', 
                        horizontalalignment='center',
                        verticalalignment='baseline',
                        fontsize = 15)
        ax.set_title(title)

def var2break(df, title, xlab, ylab, ylimtop, ylimbot):
        import matplotlib.pyplot as plt
        x_vals = df.index.nunique()
        t_c = df.sum(axis = 1)
        ylim_bot = ylimbot[1]
        fig, (ax1,ax2) = plt.subplots(2,1,figsize=(12,8),sharex = True, facecolor = 'w')
        df.plot(kind='bar', stacked=True, rot=0, ax = ax1)
        df.plot(kind='bar', stacked=True, rot=0, 
                xlabel=xlab, ylabel=ylab, ax = ax2)
        ax1.axes.xaxis.set_visible(False)
        ax2.yaxis.set_ticks_position('both')
        ax2.secondary_yaxis("right")
        ax1.set_ylim(ylimtop)
        ax2.set_ylim(ylimbot)
        ax1.spines['bottom'].set_visible(False)
        ax2.spines['top'].set_visible(False)
        ax1.xaxis.tick_bottom() 
        ax1.tick_params(labeltop=False)
        ax2.xaxis.tick_bottom()
        ax1.legend(fontsize=12)
        ax2.get_legend().remove()
        ax1.set_ylabel("")
        ax2.yaxis.set_label_coords(-0.05,1)

        for i in range(0,x_vals):
                if i == 0:
                        ax1.text(i, t_c[i], f'{t_c[i]:.0f}', horizontalalignment='center', verticalalignment='baseline', fontsize = 15)
                        ax1.text(i, t_c[i]-df[0][i]/2, f'{df[0][i]:.0f}\n{df[0][i] / t_c[i] * 100.0:.1f} %', horizontalalignment='center', verticalalignment='center')
                        ax2.text(i, ylim_bot/2, f'{df[1][i]:.0f}\n{df[1][i] / t_c[i] * 100.0:.1f} %', horizontalalignment='center', verticalalignment='center')
                else:
                        ax2.text(i, df[1][i]/2, f'{df[1][i]:.0f}\n{df[1][i] / t_c[i] * 100.0:.1f} %', horizontalalignment='center', verticalalignment='center')
                        ax2.text(i, t_c[i]-df[0][i]/2, f'{df[0][i]:.0f}\n{df[0][i] / t_c[i] * 100.0:.1f} %', horizontalalignment='center', verticalalignment='center')
                        ax2.text(i, t_c[i], f'{t_c[i]:.0f}', horizontalalignment='center', verticalalignment='baseline', fontsize = 15)
                kwargs = dict(transform=ax2.transAxes, color='k', clip_on=False)
                ax2.plot((-.01,0.01), (1.05,1.15), **kwargs)
                ax2.plot((-.01,0.01), (1.0,1.1), **kwargs)
                ax2.plot((0.99,1.01), (1.05,1.15), **kwargs)
                ax2.plot((0.99,1.01), (1.0,1.1), **kwargs)
                ax1.set_title(title)

def var2break2(df, title, var1, var2, ind1, ind2, ind1eq, ind2eq, xlab, ylab, ylimtop1, ylimbot1, ylimtop2, ylimbot2):
        df1 = totalby(df[df[ind1] == ind1eq], var1, var2)
        x_vals1 = df1.index.nunique()
        t_c = df1.sum(axis = 1)
        ylim_bot1 = ylimbot1[1]
        fig, ((ax1,ax3), (ax2, ax4)) = plt.subplots(2,2,figsize=(20,12),sharex = True, facecolor = 'w')
        df1.plot(kind='bar', stacked=True, rot=0, ax = ax1)
        df1.plot(kind='bar', stacked=True, rot=0,
                xlabel=xlab, ylabel=ylab, ax = ax2)
        ax1.axes.xaxis.set_visible(False)
        ax2.yaxis.set_ticks_position('both')
        ax2.secondary_yaxis("right")
        ax1.set_ylim(ylimtop1)
        ax2.set_ylim([0,ylim_bot1])
        ax1.spines['bottom'].set_visible(False)
        ax2.spines['top'].set_visible(False)
        ax1.xaxis.tick_bottom() 
        ax1.tick_params(labeltop=False)
        ax2.xaxis.tick_bottom()
        ax1.legend(fontsize=12)
        ax2.get_legend().remove()
        ax1.set_ylabel("")
        ax2.yaxis.set_label_coords(-0.05,1)

        for i in range(0,x_vals1):
                if i == 0:
                        ax1.text(i, t_c[i], 
                                f'{t_c[i]:.0f}', 
                                horizontalalignment='center',
                                verticalalignment='baseline',
                                fontsize = 15)
                        ax1.text(i, t_c[i]-df1[0][i]/2, 
                                f'{df1[0][i]:.0f}\n{df1[0][i] / t_c[i] * 100.0:.1f} %', 
                                horizontalalignment='center',
                                verticalalignment='center')
                        ax2.text(i, ylim_bot1/2, 
                                f'{df1[1][i]:.0f}\n{df1[1][i] / t_c[i] * 100.0:.1f} %', 
                                horizontalalignment='center',
                                verticalalignment='center')
                else:
                        ax2.text(i, df1[1][i]/2, 
                                f'{df1[1][i]:.0f}\n{df1[1][i] / t_c[i] * 100.0:.1f} %', 
                                horizontalalignment='center',
                                verticalalignment='center')
                        ax2.text(i, t_c[i]-df1[0][i]/2, 
                                f'{df1[0][i]:.0f}\n{df1[0][i] / t_c[i] * 100.0:.1f} %', 
                                horizontalalignment='center',
                                verticalalignment='center')
                        ax2.text(i, t_c[i], 
                                f'{t_c[i]:.0f}', 
                                horizontalalignment='center',
                                verticalalignment='baseline',
                                fontsize = 15)
        kwargs = dict(transform=ax2.transAxes, color='k', clip_on=False)
        ax2.plot((-.01,0.01), (1.05,1.15), **kwargs)
        ax2.plot((-.01,0.01), (1.0,1.1), **kwargs)
        ax2.plot((0.99,1.01), (1.05,1.15), **kwargs)
        ax2.plot((0.99,1.01), (1.0,1.1), **kwargs)
        if ind1 == 'class':
                ax1.set_title(title + ' class')
        else:
                ax1.set_title(title + ' ' + ind1eq + '') 

        ### df2
        df2 = totalby(df[df[ind2] == ind2eq], var1, var2)
        x_vals2 = df2.index.nunique()
        t_c = df2.sum(axis = 1)
        ylim_bot2 = ylimbot2[1]
        df2.plot(kind='bar', stacked=True, rot=0, ax = ax3)
        df2.plot(kind='bar', stacked=True, rot=0, 
                 xlabel=xlab, ylabel=ylab, ax = ax4)
        ax3.axes.xaxis.set_visible(False)
        ax4.yaxis.set_ticks_position('both')
        ax4.secondary_yaxis("right")
        ax3.set_ylim(ylimtop2)
        ax4.set_ylim([0,ylim_bot2])
        ax3.spines['bottom'].set_visible(False)
        ax4.spines['top'].set_visible(False)
        ax3.xaxis.tick_bottom() 
        ax3.tick_params(labeltop=False)
        ax4.xaxis.tick_bottom()
        ax3.legend(fontsize=12)
        ax4.get_legend().remove()
        ax3.set_ylabel("")
        ax4.yaxis.set_label_coords(-0.05,1)

        for i in range(0,x_vals2):
                if i == 0:
                        ax3.text(i, t_c[i]-df2[0][i]/2, 
                                f'{df2[0][i]:.0f}\n{df2[0][i] / t_c[i] * 100.0:.1f} %', 
                                horizontalalignment='center',
                                verticalalignment='center')
                        ax4.text(i, ylim_bot2/2, 
                                f'{df2[1][i]:.0f}\n{df2[1][i] / t_c[i] * 100.0:.1f} %', 
                                horizontalalignment='center',
                                verticalalignment='center')
                        ax3.text(i, t_c[i], 
                        f'{t_c[i]:.0f}', 
                        horizontalalignment='center',
                        verticalalignment='baseline',
                        fontsize = 15)
                else:
                        if df2[1][i] > 0:
                                ax4.text(i, df2[1][i]/2, 
                                        f'{df2[1][i]:.0f}\n{df2[1][i] / t_c[i] * 100.0:.1f} %', 
                                        horizontalalignment='center',
                                        verticalalignment='center')
                        if df2[0][i] > 0:
                                ax4.text(i, t_c[i]-df2[0][i]/2, 
                                        f'{df2[0][i]:.0f}\n{df2[0][i] / t_c[i] * 100.0:.1f} %', 
                                        horizontalalignment='center',
                                        verticalalignment='center')
                                ax4.text(i, t_c[i], 
                                f'{t_c[i]:.0f}', 
                                horizontalalignment='center',
                                verticalalignment='baseline',
                                fontsize = 15)
        kwargs = dict(transform=ax4.transAxes, color='k', clip_on=False)
        ax4.plot((-.01,0.01), (1.05,1.15), **kwargs)
        ax4.plot((-.01,0.01), (1.0,1.1), **kwargs)
        ax4.plot((0.99,1.01), (1.05,1.15), **kwargs)
        ax4.plot((0.99,1.01), (1.0,1.1), **kwargs)
        if ind2eq == 'Only test':
                ax3.set_title(title + ' test prep')
        else:
                ax3.set_title(title + ' ' + ind2eq + '') 
```
